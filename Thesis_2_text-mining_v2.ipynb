```python
import numpy as np
import pandas as pd
import re
import matplotlib
import csv

%matplotlib inline
```

```python
df = pd.read_csv('all_reviews_v3.csv', index_col=0)
```

```python
def find_nausea(comment):
    m = re.search('nausée|nausee', comment)
    return m is not None
df['nausea'] = df['comment'].fillna("").apply(find_nausea)
```

```python
def find_fainting(comment):
    m = re.search('evanoui|évanoui|tomber dans les pommes|tombe dans les pommes|tombé dans les pommes', comment)
    return m is not None
df['fainting'] = df['comment'].fillna("").apply(find_fainting)
```

```python
def find_headache(comment):
    m = re.search('migraine|mal a la tete|mal de tete|maux de tete|mal à la tête|maux de tête|mal de tête|mal a la tête|mal à la tete|céphalée', comment)
    return m is not None
df['headache'] = df['comment'].fillna("").apply(find_headache)
```

```python
def find_vomit(comment):
    m = re.search('vomi', comment)
    return m is not None
df['vomit'] = df['comment'].fillna("").apply(find_vomit)
```

```python
def find_depression(comment):
    m = re.search('deprim|déprim', comment)
    return m is not None
df['depression'] = df['comment'].fillna("").apply(find_depression)
```

```python
def find_fatigue(comment):
    m = re.search('fatigue', comment)
    return m is not None
df['fatigue'] = df['comment'].fillna("").apply(find_fatigue)
```

```python
def find_stomache_ache(comment):
    m = re.search("maux de ventre|mal de ventre|maux d'estomac|mal a l'estomac",comment)
    return m is not None
df['stomache_ache'] = df['comment'].fillna("").apply(find_stomache_ache)
```

```python
def find_digestive_issues(comment):
    m = re.search('trouble digestif|troubles digestif|problemes digestifs|problèmes digestifs', comment)
    return m is not None
df['digestive_issues'] = df['comment'].fillna("").apply(find_digestive_issues)
```

```python
def find_constipation(comment):
    m = re.search('constip', comment)
    return m is not None
df['constipation'] = df['comment'].fillna("").apply(find_constipation)
```

```python
def find_acne(comment):
    m = re.search('acne|acné|bouton', comment)
    return m is not None
df['acne'] = df['comment'].fillna("").apply(find_acne)
```

```python
def find_hallucinations(comment):
    m = re.search('hallu', comment)
    return m is not None
df['hallucinations'] = df['comment'].fillna("").apply(find_hallucinations)
```

```python
def find_dizziness(comment):
    m = re.search('vertige', comment)
    return m is not None
df['dizziness'] = df['comment'].fillna("").apply(find_dizziness)
```

```python
df['total'] = df['comment'].notnull()
```

```python
hi = pd.pivot_table(df, values=["nausea","fainting","headache", "vomit", "depression", "fatigue", "stomache_ache",
                                "constipation", "acne", "digestive_issues", "total"], 
                    index="medicine", aggfunc=np.sum).reset_index()
hi
```
